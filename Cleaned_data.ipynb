```python
df=pd.read_csv("D:\games.csv\games.csv")
```

```python
#converting release dates to date time
df["Release date"]=pd.to_datetime(df["Release date"],errors="coerce")
```

```python
#reviewing the null values
df.isna().sum()
```

```python
df.loc[df['About the game'].isna(),"Name"]
```

```python
#Since most of the null game descriptions are playsets with no decscriptions, we can replace them with suitable text
df["Name"].fillna("", inplace=True)
df.loc[df["Name"].str.contains('Playtest', na=False), 'About the game'] = "this is a playtest game is the process by which a game designer tests a new game for bugs and design flaws before releasing it to market."
```

```python
df.loc[df['About the game'].isna(),"Name"]
```

```python
# exploring other decsriotions, most games are beta/alpha and such so we will replace with suitable texts
keyword_descriptions = {
    'Beta': 'this game is beta and still under testing',
    'Alpha': 'this game is Alpha and still under testing',
    'beta': 'this game is beta and still under testing',
    'BETA': 'this game is beta and still under testing',
    'Test': 'this game is still under testing',
    'playtest': 'this game is still under testing',
    'SDK': 'Software Development Kit of the game',
    'Demo': 'this game is Demo and still under testing',
    'Server': 'this is a Server for the game',
    'Editor': 'this is an Editor for the game'
}

# Iterate over the rows of the DataFrame
for index, row in df.iterrows():
    if pd.isnull(row['About the game']):
        # Iterate over the keywords and descriptions
        for keyword, description in keyword_descriptions.items():
            # Check if the keyword is in the game name
            if keyword in row['Name']:
                df.at[index, 'About the game'] = description
                break  # Exit the loop once a match is found
        else:
            df.at[index, 'About the game'] = 'this game does not have a description'
```

```python
count=0
for index, row in df.iterrows():
    if pd.isnull(row['Screenshots']):
        if isinstance(row['Name'], str) and 'Playtest' in row['Name']:
            count += 1
        else:
            continue
print(count)
```

```python
#Since most of the null values of screenshots are of playtest (in development), we can drop them
df.dropna(subset=['Screenshots'], inplace=True)
```

```python
#checking null genres of in development games
keywords=['Playtest', 'Alpha', 'beta', 'BETA', 'Test', 'playtest', 'SDK', 'Demo', 'Server', 'Editor', 'Beta']
count=0
for index, row in df.iterrows():
    if pd.isnull(row["Genres"]) and not pd.isnull(row["Name"]):
        if any(keyword in row["Name"] for keyword in keywords):
            count+=1
print(count)
```

```python
#Since most genres are of ongoing development games and noy playable, we will replace their nulls with suitable text
genre_mapping = {
    'Playtest': 'Playtest game is not playayble',
    'Alpha': 'Alpha game is not playayble',
    'beta': 'Beta game is not playayble',
    'BETA': 'Beta game is not playayble',
    'Test': 'Test game is not playayble',
    'playtest': 'Playtest game is not playayble',
    'SDK': 'Software Development Kit of the game is not playayble',
    'Demo': 'Demo game is not playayblee',
    'Server': 'Server of a game is not playayble',
    'Editor': 'Editor of a game is not playayble',
    'Beta': 'Beta game is not playayble'
}

for index, row in df.iterrows():
    if pd.isnull(row['Genres']):
        if isinstance(row['Name'], str):  
            for keyword, genre in genre_mapping.items():
                if keyword in row['Name']:
                    df.at[index, 'Genres'] = genre
                    break
            else:
                df.at[index, 'Genres'] = 'no genre available'
        else:
            df.at[index, 'Genres'] = 'no genre available'
```

```python
#checking null support emails of in development games
keywords=['Playtest', 'Alpha', 'beta', 'BETA', 'Test', 'playtest', 'SDK', 'Demo', 'Server', 'Editor', 'Beta']
count=0
for index, row in df.iterrows():
    if pd.isnull(row["Support email"]) and not pd.isnull(row["Name"]):
        if any(keyword in row["Name"] for keyword in keywords):
            count+=1
print(count)            
```

```python
#Since most support emails are of ongoing development games, we will replace their nulls with suitable text
support_email_mapping = {
    'Playtest': 'Playtest game no Support email available',
    'Alpha': 'Alpha game no Support email available',
    'beta': 'Beta game no Support email available',
    'BETA': 'Beta game no Support email available',
    'Test': 'Test game no Support email available',
    'playtest': 'Playtest game no Support email available',
    'SDK': 'Software Development Kit of the game no Support email available',
    'Demo': 'Demo game no Support email available',
    'Server': 'Server of a game no Support email available',
    'Editor': 'Editor of a game no Support email available',
    'Beta': 'Beta game no Support email available'
}

for index, row in df.iterrows():
    if pd.isnull(row['Support email']):
        if isinstance(row['Name'], str):  
            for keyword, email_message in support_email_mapping.items():
                if keyword in row['Name']:
                    df.at[index, 'Support email'] = email_message
                    break
            else:
                df.at[index, 'Support email'] = 'no Support email available'
        else:
            df.at[index, 'Support email'] = 'no Support email available'
```

```python
#Checking null values of Movies
keywords = ['Playtest', 'Alpha', 'beta', 'BETA', 'Test', 'playtest', 'SDK', 'Demo', 'Server', 'Editor', 'Beta']
count = 0
for index, row in df.iterrows():
    if pd.isnull(row['Movies']) and not pd.isnull(row['Name']): 
        if any(keyword in row['Name'] for keyword in keywords):  
            count += 1
print(count)
```

```python
#Since most of the movies null values are of in development games, we can drop them
df.dropna(subset=['Movies'], inplace=True)
```

```python
#checking developers column null
df.loc[df['Developers'].isna()]
```

```python
#checking publishers column null
df.loc[df['Publishers'].isna()]
```

```python
#deleting notes, score rank, Metacritic url, website, Support url columns because they have alot of null values more than 50% and irrevelant
df = df.drop('Notes', axis=1)
df = df.drop('Score rank', axis=1)
df = df.drop('Metacritic url', axis=1)
df = df.drop('Support url', axis=1)
df = df.drop('Website', axis=1)
```

```python
count = 0
for index, row in df.iterrows():
    if(pd.isnull(row['Publishers'])):
        if(pd.isnull(row['Developers'])):
            continue
        else:
            print(row['Developers'])
            count += 1
print(count)
```

```python
#Since 289 null publishers have developers, i will named the publisher same as developer
for index, row in df.iterrows():
    if(pd.isnull(row['Publishers'])):
        if(pd.isnull(row['Developers'])):
            continue
        else:
            df.at[index, 'Publishers'] = df.at[index, 'Developers']
```

```python
count = 0
for index, row in df.iterrows():
    if(pd.isnull(row['Developers'])):
        if(pd.isnull(row['Publishers'])):
            continue
        else:
            print(row['Publishers'])
            count += 1
print(count)
```

```python
#Since 58 null developers have publishers, i will named the developers same as publishers
for index, row in df.iterrows():
    if(pd.isnull(row['Developers'])):
        if(pd.isnull(row['Publishers'])):
            continue
        else:
            df.at[index, 'Developers'] = df.at[index, 'Publishers']
```

```python
#most of the tags are similar to genres so ill replace the tags by genre
for index, row in df.iterrows():
    if(pd.isnull(row['Tags'])):
        df.at[index, 'Tags'] = df.at[index, 'Genres']
```

```python
#the windows, linux and MAC are in three different columns, ill combine them by creating a column called OS
for index, row in df.iterrows():
    if row["Windows"]:
        os = "Windows"
        if row["Mac"]:
            os += ", Mac"
        if row["Linux"]:
            os += ", Linux"
    elif row["Mac"]:
        os = "Mac"
        if row["Linux"]:
            os += ", Linux"
    elif row["Linux"]:
        os = "Linux"
    else:
        os = "Unknown"
    
    df.at[index, 'OS'] = os
```
